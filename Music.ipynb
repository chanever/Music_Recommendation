### 3. 함수 find_music_on_yt

```python
def find_music_on_yt(c):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get(c)

    #웹이 로드될때까지 기다린다.
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,"content")))

    html_page = browser.page_source
    browser.quit()


    
    #로드한 html파일을 보기 좋게 정리해준다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 클래스이름을 가진 a태그를 가져온다.

    prodList = soup.find_all("div") #div태그로 들어있는 html코드를 가져온다.
    mlist = []
    for prod in prodList: #list형식으로 되어있는 html코드들을 하나씩 보며 그 코드 안에서 원하는 클래스를 가진 a태그를 찾는다.
        mlist.extend(prod.find_all("a", {"class":"yt-simple-endpoint style-scope ytd-playlist-video-renderer"}))
        
    mlist2 = []
    for m in mlist:
        mlist2.append(m['href']) #원하는 href속성을 가져오기 위해 href를 검색한다.
    
    return mlist2
```

### 4. 함수 random_three_music

```python
def random_three_music(mlist):
    three_music=[]
    while True:        
        music=random.choice(mlist)
        if music not in three_music:
           three_music.append(music)
        if len(three_music)==3:
            break
    return three_music
```

```python
playlist_com=os.getcwd() #현재 작업중인 폴더경로를 뽑아옵니다.
def search_on_sc(a):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get('https://soundcloud.com/search/sets?q='+a)

    #웹이 로드될때까지 기다린다.
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.CLASS_NAME,"soundTitle__usernameTitleContainer")))

    html_page = browser.page_source
    browser.quit()



    #로드한 html파일을 보기 좋게 정리해준다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 클래스이름을 가진 div를 가져온다.

    prodList = soup.find_all("a", {'class':'soundTitle__title sc-link-dark'})
    links = [data['href'] for data in prodList]
    with open(playlist_com + '/' + a +'사클재생목록저장.txt','a') as 저장파일:
         for link in links:
            저장파일.write(link+'\n')
```

```python
!pip install selenium
import requests
from selenium import webdriver
from bs4 import BeautifulSoup
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.support import expected_conditions as EC 
from selenium.webdriver.common.by import By
import random
import webbrowser
import os



playlist_com=os.getcwd() #현재 작업중인 폴더경로를 뽑아옵니다.
def search_on_yt(a):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get('https://www.youtube.com/results?search_query='+a)

    #웹이 로드될때까지 기다립니다.
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,"view-more")))

    html_page = browser.page_source
    browser.quit()



    #로드한 html파일을 보기 좋게 정리해줍니다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 원하는 클래스의 데이터들을 가져와줍니다.

    prodList = soup.find_all("yt-formatted-string", {'id':'view-more'})
    links = [data.contents[0]['href'] for data in prodList]
    with open(playlist_com + '/' + a +'유튭재생목록저장.txt','a') as 저장파일:
         for link in links:
            저장파일.write(link+'\n')


        
#2 재생목록 랜덤적으로 뽑기

def find_playlist_on_yt(a):
    with open(a+"유튭재생목록저장.txt", 'r') as f:
        리스트변환=f.readlines()
    url추출=random.choice(리스트변환).strip()
    return 'https://www.youtube.com'+url추출
    
        
        
        
    
    
#3 재생목록에 들어가 노래링크 찾기

def find_music_on_yt(c):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get(c)

    #웹이 로드될때까지 기다린다.
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,"content")))

    html_page = browser.page_source
    browser.quit()


    
    #로드한 html파일을 보기 좋게 정리해준다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 클래스이름을 가진 div태그를 가져옵니다.

    prodList = soup.find_all("div") #div태그로 들어있는 html코드를 가져온다.
    mlist = []
    for prod in prodList: #list형식으로 되어있는 html코드들을 하나씩 보며 그 코드 안에서 원하는 클래스를 가진 a태그를 찾는다.
        mlist.extend(prod.find_all("a", {"class":"yt-simple-endpoint style-scope ytd-playlist-video-renderer"}))
        
    mlist2 = []
    for m in mlist:
        mlist2.append(m['href']) #원하는 href속성을 가져오기 위해 href를 검색한다.
    
    return mlist2
    
    
#4 노래 찾은후 3가지의 랜덤 노래 추출 

def random_three_music(mlist):
    three_music=[]
    while True:        
        music=random.choice(mlist)
        if music not in three_music:
           three_music.append(music)
        if len(three_music)==3:
            break
    return three_music


#5 유튜브창에서 노래 띄우기

def browse_music_on_yt(three_music):
    for music in three_music:
        webbrowser.open('https://www.youtube.com'+music,new=2) #three music에 있는 각 url을 완성 후 웹브라우저로 실행
    

    

    
playlist_com=os.getcwd() #현재 작업중인 폴더경로를 뽑아옵니다.
def search_on_sc(a):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get('https://soundcloud.com/search/sets?q='+a)

    #웹이 로드될때까지 기다립니다.
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.CLASS_NAME,"soundTitle__usernameTitleContainer")))

    html_page = browser.page_source
    browser.quit()



    #로드한 html파일을 보기 좋게 정리해준다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 원하는 클래스의 데이터들을 가져와줍니다.

    prodList = soup.find_all("a", {'class':'soundTitle__title sc-link-dark'})
    links = [data['href'] for data in prodList]
    with open(playlist_com + '/' + a +'사클재생목록저장.txt','a') as 저장파일:
         for link in links:
            저장파일.write(link+'\n')


        
#2 재생목록 랜덤적으로 뽑기

def find_playlist_on_sc(a):
    with open(a+"사클재생목록저장.txt", 'r') as f:
        리스트변환=f.readlines()
    url추출=random.choice(리스트변환).strip()
    return 'https://soundcloud.com'+url추출
    
        
        
        
    
    
#3 재생목록에 들어가 노래링크 찾기

def find_music_on_sc(c):
    options = webdriver.ChromeOptions()
    options.add_argument('headless')
    options.add_argument('window-size=1920x1080')
    options.add_argument("disable-gpu")

    browser = webdriver.Chrome('C:/chromedriver.exe',chrome_options=options)
    browser.get(c)

    #웹이 로드될때까지 기다립니다
    WebDriverWait(browser,10).until(EC.visibility_of_element_located((By.ID,"content")))

    html_page = browser.page_source
    browser.quit()


    
    #로드한 html파일을 보기 좋게 정리해준다.
    soup = BeautifulSoup(html_page, 'html.parser')



    #정리된 html파일에서 해당 클래스이름을 가진 a태그를 가져온다.

    prodList = soup.find_all("div", {"class":"trackItem__content sc-truncate"}) #div태그로 들어있는 html코드를 가져온다.
    
    mlist2 = [data.contents[5]['href'] for data in prodList] # pprodList는 div의 원하는 클래스를 뽑았고 그 데이터들은 배열의 형태로 구분되어있다.
    #따라서 contents가 뽑은 리스트 형태의 데이터에서 5번째 객체에서 href 추출함.
    
    
    return mlist2
    
    
#4 노래 찾은후 3가지의 랜덤 노래 추출 

def random_three_music(mlist):
    three_music=[]
    while True:        
        music=random.choice(mlist)
        if music not in three_music:
           three_music.append(music)
        if len(three_music)==3:
            break
    return three_music


#5  사운드 클라우드에서 노래 띄우기

def browse_music_on_sc(three_music):
    for music in three_music:
        webbrowser.open('https://soundcloud.com'+music,new=2) #three music에 있는 각 url을 완성 후 웹브라우저로 실행
#메인코드
우울문구=['우울한일이 있으신가봐요.. 이럴수록 힘내야져!',
      '이노래듣고 기운내요 ㅠㅠ  ','이노래는 너의 입맛에 맞을거야!']

슬픔문구=['슬픈일이 있으신가봐요.. 이럴수록 힘내야져!',
      '이노래듣고 기운내요 ㅠㅠ  ','이노래는 너의 입맛에 맞을거야!' , '노래 들으면서 좋은생각, 행복했던 일을 생각해봐!']

행복문구=[ '행복한일이 있으신가봐요! 이노래듣고 오늘 하루도 파이팅!','그 행복이 계속되길!!','행복하면 크게 웃어보자!',
      '이노래듣고 더 행복해져 ㅎㅎ ','이노래는 너의 입맛에 맞을거야!']

기쁨문구=['좋은일이 있으신가봐요! 이노래듣고 오늘 하루도 파이팅!','긍정적인 모습 아주좋아!','기쁘면 당연히 이노래지!',
      '이노래듣고 더 기뻐져라 ㅎㅎ ','이노래는 너의 입맛에 맞을거야!']

화남문구=['화난일이 있으신가봐요.. 이럴수록 진정!!',
      '이노래듣고 기운내요 ㅠㅠ  ','이노래는 너의 입맛에 맞을거야!' , '노래 들으면서 좋은생각, 행복했던 일을 생각해봐!']

if __name__=='__main__':
    while True:
        b=int(input('''음악 플랫폼 둘중에 하나 골라봐!
        1)유튜브
        2)사운드 클라우드
        '''))
        
        
        
        
        a=int(input('''노래 장르를 선택해!
        1)우울 
        2)슬픔
        3)기쁨
        4)행복
        5)화남
        '''))
        if a == 1:
            if b==1:
            # os.listdir 해당경로에 있는 파일을 리스트로 만듬
            
                if"depressing+songs유튭재생목록저장.txt" not in os.listdir('/'):
                    search_on_yt("depressing+songs")
                print(random.choice(우울문구))
                browse_music_on_yt(random_three_music(find_music_on_yt(find_playlist_on_yt("depressing+songs"))))
                break
            elif b==2:
                if"depressing사클재생목록저장.txt" not in os.listdir('/'):
                    search_on_sc("depressing")
                print(random.choice(우울문구))
                browse_music_on_sc(random_three_music(find_music_on_sc(find_playlist_on_sc("depressing"))))
                break
            else:
                print("플랫폼부터 다시고르세요.")
            
        elif a == 2:
            if b==1:
                
                if"sad+songs유튭재생목록저장.txt" not in os.listdir('/'):
                
                    search_on_yt("sad+songs")
                print(random.choice(슬픔문구))
                browse_music_on_yt(random_three_music(find_music_on_yt(find_playlist_on_yt("sad+songs"))))
                break
            elif b==2:
                if"sad사클재생목록저장.txt" not in os.listdir('/'):
                    search_on_sc("sad")
                print(random.choice(우울문구))
                browse_music_on_sc(random_three_music(find_music_on_sc(find_playlist_on_sc("sad"))))
                break
            else:
                print("플랫폼부터 다시고르세요.")
        elif a == 3:
            if b==1:
                if"exciting+songs유튭재생목록저장.txt" not in os.listdir('/'):
                    search_on_yt("exciting+songs")
                print(random.choice(기쁨문구))
                browse_music_on_yt(random_three_music(find_music_on_yt(find_playlist_on_yt("exciting+songs"))))
                break
                
            elif b==2:
                if"exciting+사클재생목록저장.txt" not in os.listdir('/'):
                    search_on_sc("exciting+")
                print(random.choice(우울문구))
                browse_music_on_sc(random_three_music(find_music_on_sc(find_playlist_on_sc("exciting+"))))
                break
            else:
                print("플랫폼부터 다시고르세요.")
        elif a == 4:
            if b==1:
                if"happy+songs유튭재생목록저장.txt" not in os.listdir('/'):
                    
                    search_on_yt("happy+songs")
                print(random.choice(행복문구))
                browse_music_on_yt(random_three_music(find_music_on_yt(find_playlist_on_yt("happy+songs"))))
                break
                
            elif b==2:
                if"happy사클재생목록저장.txt" not in os.listdir('/'):
                    search_on_sc("happy")
                print(random.choice(우울문구))
                browse_music_on_sc(random_three_music(find_music_on_sc(find_playlist_on_sc("happy"))))
                break
            else:
                print("플랫폼부터 다시고르세요.")
                
        elif a == 5:
            if b==1:
                
                if"chill+songs유튭재생목록저장.txt" not in os.listdir('/'):
                    search_on_yt("chill+songs")
                print(random.choice(화남문구))
                browse_music_on_yt(random_three_music(find_music_on_yt(find_playlist_on_yt("chill+songs"))))
                break
                
            elif b==2:
                if"chill사클재생목록저장.txt" not in os.listdir('/'):
                    search_on_sc("chill")
                print(random.choice(우울문구))
                browse_music_on_sc(random_three_music(find_music_on_sc(find_playlist_on_sc("chill"))))
                break
            else:
                print("플랫폼부터 다시고르세요.")
        else:
            print("1번부터 5번중 하나를 선택해주세요.")
```
